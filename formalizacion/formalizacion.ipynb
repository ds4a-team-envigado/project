```python
formalizacion_df.drop(columns=['Nombre del solicitante', 'Telefono solicitante', 'E-mail solicitante'])
formalizacion_df[['Total beneficiarios', 'Hombres', 'Mujeres']].sum()
```

```python
formalizacion_df['Total beneficiarios'].fillna(formalizacion_df['Total beneficiarios'].quantile(0.50), inplace=True)
formalizacion_df['Hombres'].fillna(formalizacion_df['Total beneficiarios']/2, inplace=True)
formalizacion_df['Mujeres'].fillna(formalizacion_df['Total beneficiarios']/2, inplace=True)
formalizacion_df
```

```python
formalizacion_df.groupby(by='Departamento').agg({'Total beneficiarios':['count','mean'], 
                                                 'Hombres':['sum','mean'],
                                                 'Mujeres':['sum','mean']})
```
