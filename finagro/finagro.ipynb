```python
import numpy as np
import pandas as pd
```

Las bases de datos se encuentran en el drive.

```python
df17 = pd.read_csv('finagro_2017.csv')
df18 = pd.read_csv('finagro_2018.csv')
df19 = pd.read_csv('finagro_2019.csv')

[df17.shape, df18.shape, df19.shape]
```

```python
df19.columns
```

```python
prod17 = df17['Linea de Produccion'].value_counts().sort_index()
prod18 = df18['Linea de Produccion'].value_counts().sort_index()
prod19 = df19['Linea de Produccion'].value_counts().sort_index()
prod = pd.DataFrame(zip(prod19.index, prod17.values, prod18.values, prod19.values),
             columns = ['Linea - Produccion', '2017', '2018', '2019'])
prod['Tipo'] = prod['Linea - Produccion'].apply(lambda x: str(x)[-4:])
prod.groupby(['Tipo', 'Linea - Produccion'])['2017', '2018', '2019'].sum()
```

```python
dep17 = df17['Departamento Inversion'].value_counts().sort_index()
dep18 = df18['Departamento Inversion'].value_counts().sort_index()
dep19 = df19['Departamento Inversion'].value_counts().sort_index()
dep = pd.DataFrame(zip(dep19.index, dep17.values, dep18.values, dep19.values),
             columns = ['departamento', '2017', '2018', '2019'])
dep
```

```python
destino17 = df17['Destino de Credito'].value_counts().sort_index().index
destino18 = df18['Destino de Credito'].value_counts().sort_index().index
destino19 = df19['Destino de Credito'].value_counts().sort_index().index
destino = pd.concat([pd.DataFrame(destino17), pd.DataFrame(destino18), pd.DataFrame(destino19)])[0].value_counts().sort_index()
destino
```
