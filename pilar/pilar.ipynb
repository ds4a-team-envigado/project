```python
gp_by_department = df.groupby('Departamento')['BP'].count().sort_values(ascending=False).to_frame().reset_index() 
gp_by_department

gp_by_valor = df.groupby('Departamento')['Valor cofinanciado ADR'].sum().sort_values(ascending=False).sort_values(ascending=False).to_frame().reset_index() 
gp_by_valor



gp_b = df.groupby('Departamento')['Total beneficiarios'].sum().sort_values(ascending=False).sort_values(ascending=False).to_frame().reset_index() 
gp_b

df_merged = pd.merge(gp_by_department, gp_b)

df_merged2 = pd.merge(df_merged, gp_by_valor)
```
